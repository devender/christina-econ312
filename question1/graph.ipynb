```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates for the vertical line
x_vertical = [2, 2]  # X-coordinates (same value for a vertical line)
y_vertical = [0, 10]  # Y-coordinates (different values to create a vertical span)

# Define the range and equation for the parabola
x_parabola = np.linspace(0, 4, 100)  # Generates 100 points between 0 and 4
y_parabola = x_parabola**2  # Parabolic function (y = x^2)

# Plot the vertical line
plt.plot(x_vertical, y_vertical, color='blue', label='Vertical Line at x = 2')

# Plot the parabola
plt.plot(x_parabola, y_parabola, color='red', label='Parabola y = x^2')

# Adding titles, labels, and a legend
plt.title('Vertical Line and Parabola')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
